```python
import pandas as pd
import requests
```

```python
url = "https://archive-api.open-meteo.com/v1/archive"

params = {
    "latitude": 48.8566,
    "longitude": 2.3522,
    "start_date": "2020-01-01",
    "end_date": "2023-12-31",
    "daily": "temperature_2m_max,temperature_2m_min,temperature_2m_mean,precipitation_sum,rain_sum,showers_sum,wind_speed_10m_max,wind_gusts_10m_max,relative_humidity_2m_mean,cloud_cover_mean,sunshine_duration,shortwave_radiation_sum",
    "timezone": "Europe/Paris"
}

response = requests.get(url, params=params)
data = response.json()
```

```python
cities = {
    "Paris": (48.8566, 2.3522),
    "Lyon": (45.7640, 4.8357),
    "Marseille": (43.2965, 5.3698),
    "Lille": (50.6292, 3.0573),
    "Bordeaux": (44.8378, -0.5792)
}

all_data = []

for city, coords in cities.items():
    
    params = {
        "latitude": coords[0],
        "longitude": coords[1],
        "start_date": "2014-01-01",
        "end_date": "2023-12-31",
        "daily": "temperature_2m_max,temperature_2m_min,temperature_2m_mean,precipitation_sum,rain_sum,showers_sum,wind_speed_10m_max,wind_gusts_10m_max,relative_humidity_2m_mean,cloud_cover_mean,sunshine_duration,shortwave_radiation_sum",
        "timezone": "Europe/Paris"
    }
    
    response = requests.get(url, params=params)
    data = response.json()
    
    df = pd.DataFrame(data["daily"])
    df["city"] = city
    
    all_data.append(df)

weather_data = pd.concat(all_data)

weather_data.head()
```

```python
weather_data["time"] = pd.to_datetime(weather_data["time"])
```

```python
weather_data["month"] = weather_data["time"].dt.month
```

```python
weather_data["year"] = weather_data["time"].dt.year
```

```python
def get_season(month):
    if month in [12,1,2]:
        return "winter"
    elif month in [3,4,5]:
        return "spring"
    elif month in [6,7,8]:
        return "summer"
    else:
        return "autumn"

weather_data["season"] = weather_data["month"].apply(get_season)
```
